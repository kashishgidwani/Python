```python
import pandas as pd
df1=pd.DataFrame([[1,2],[3,4]],columns=['a','b'])
df2=pd.DataFrame([[5,6],[7,8]],columns=['a','b'])
df1=df1.append(df2)
print(df1)
print("\nAfter deleting")
df1=df1.drop(0)
print(df1)
print("\nAfter adding a new column")
df1["c"]=[6,5]
print(df1)
```

```python
#create a dataframe
import pandas as pd
df=pd.DataFrame()
df["Name"]=["ankit","Ankita","yashv","Vikal"]
df["CAT1_Mark"]=[15,22,24,25]
df["CAT2_Mark"]=[18,24,21,23]
df["FAT_Mark"]=[45,50,37,49]
print(df)

#adding one more row
df.loc[len(df.index)]=["Pavan",23,23,41]
print("\n\n",df)

#adding a new column to store total marks
df["Total"]=df["CAT1_Mark"]+df["CAT2_Mark"]+df["FAT_Mark"]
print("\n\n",df)
print("\n\nSorting data according to total marks:")
df.sort_values("Total")
```
